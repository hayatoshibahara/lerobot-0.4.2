## データセットの作成

[初期設定][1]完了後、録画コマンドを実行する。

```sh
export REPO_NAME="grab-a-cube-v0"
export NUM_EPISODES=50
export TASK="Grab a cube"

lerobot-record \
    --robot.type=so101_follower \
    --robot.port=$FOLLOWER_PORT \
    --robot.id=$FOLLOWER_ID \
    --robot.cameras=$CAMERAS \
    --teleop.type=so101_leader \
    --teleop.port=$LEADER_PORT \
    --teleop.id=$LEADER_ID \
    --dataset.repo_id=$HF_USER/$REPO_NAME \
    --dataset.num_episodes=$NUM_EPISODES \
    --dataset.single_task=$TASK \
    --dataset.push_to_hub=True \
    --display_data=True \
    --play_sounds=True \
    --resume=False
```

エラーなどで中断した場合は、`--resume=True`で再開できる（`--dataset.num_episodes`は追加のエピソード数になる）。

最初から記録したい場合は、データセットディレクトリを削除する。

コマンド実行後はキーボード操作を受け付ける（`init_keyboard_listener()`）:

| キー | 動作 | 説明 |
| :--- | :--- | :--- |
| **右矢印キー** (`->`) | 次へ進む / 早期終了 | 現在のループ（エピソードの録画やリセット待機時間など）を早期終了し、次のステップに進みます。 |
| **左矢印キー** (`<-`) | 再録画 | 現在のループを中断し、直前のエピソードを破棄して、再度そのエピソードの録画をやり直します。 |
| **Escキー** | 録画終了 | データ録画セッション全体を停止し、終了処理に移行します。 |


コマンド引数の説明は`src/lerobot/scripts/lerobot_record.py`の`RecordConfig`にある。

一般的なコマンド引数（RecordConfigクラス）:

| フィールド名 | 型 | デフォルト値 | 説明 |
| :--- | :--- | :--- | :--- |
| `repo_id` | `str` | (必須) | データセットの識別子。慣例として `{hf_username}/{dataset_name}` (例: `lerobot/test`) と一致させる必要があります。 |
| `single_task` | `str` | (必須) | 録画中に行われるタスクの簡潔かつ正確な説明（例：「レゴブロックを拾って右側の箱に入れる」）。 |
| `root` | `str \| Path \| None` | `None` | データセットが保存されるルートディレクトリ（例: `dataset/path`）。 |
| `fps` | `int` | `30` | 1秒あたりのフレーム数（FPS）を制限します。 |
| `episode_time_s` | `int \| float` | `60` | 各エピソードのデータ録画を行う秒数。 |
| `reset_time_s` | `int \| float` | `60` | 各エピソードの後に環境をリセットするための秒数。 |
| `num_episodes` | `int` | `50` | 録画するエピソードの数。 |
| `video` | `bool` | `True` | データセット内のフレームを動画としてエンコードするかどうか。 |
| `push_to_hub` | `bool` | `True` | データセットをHugging Face Hubにアップロードするかどうか。 |
| `private` | `bool` | `False` | Hugging Face Hubのプライベートリポジトリにアップロードするかどうか。 |
| `tags` | `list[str] \| None` | `None` | Hub上のデータセットにタグを追加します。 |
| `num_image_writer_processes` | `int` | `0` | フレームをPNGとして保存する処理を行うサブプロセスの数。<br>0に設定するとスレッドのみを使用します。1以上に設定するとサブプロセスを使用し、各プロセスがスレッドを使用して画像を書き込みます。<br>最適なプロセス数とスレッド数はシステムに依存します。0プロセスでカメラごとに4スレッドを推奨します。<br>FPSが不安定な場合はスレッド数を調整してください。それでも不安定な場合は、1つ以上のサブプロセスの使用を試してください。 |
| `num_image_writer_threads_per_camera` | `int` | `4` | カメラごとにフレームをPNG画像としてディスクに書き込むスレッドの数。<br>スレッドが多すぎるとメインスレッドがブロックされ、テレオペレーションのFPSが不安定になる可能性があります。<br>スレッドが少なすぎるとカメラのFPSが低下する可能性があります。 |
| `video_encoding_batch_size` | `int` | `1` | 動画をバッチエンコードする前に録画するエピソードの数。<br>即時エンコード（デフォルトの動作）の場合は1、バッチエンコードの場合はそれより大きな値を設定します。 |
| `rename_map` | `dict[str, str]` | `{}` | 観測データ（画像や状態のキー）を上書きするためのリネームマップ。 |

フォロワーアームのコマンド引数（RobotConfigクラス）:

| フィールド名 | 型 | デフォルト値 | 説明 |
| :--- | :--- | :--- | :--- |
| `id` | `str \| None` | `None` | 同じタイプの異なるロボットを区別するために使用します。 |
| `calibration_dir` | `Path \| None` | `None` | キャリブレーションファイルを保存するディレクトリ。 |

| フィールド名 | 型 | デフォルト値 | 説明 |

| `id` | `str \| None` | `None` | 同じタイプの異なるテレオペレータを区別するために使用します。 |

## データセットのアップロード

データセットは`~/.cache/huggingface/lerobot/${HF_USER}/${REPO_NAME}`に保存されている。

コマンド実行後にデータセットはHuggingFaceに自動的にアップロードできるが、手動でのアップロードもできる。

```sh
huggingface-cli upload ${HF_USER}/${REPO_NAME} ~/.cache/huggingface/lerobot/${HF_USER}/${REPO_NAME} --repo-type dataset
```

## データセットの可視化

[LeRobot Dataset Visualizer][1]で確認できる。

コマンドで確認も可能。

```sh
lerobot-dataset-viz \
    --repo-id lerobot/pusht \
    --episode-index 0
```

ローカルフォルダを指定して確認もできる。

```sh
lerobot-dataset-viz \
    --repo-id lerobot/pusht \
    --root ./my_local_data_dir \
    --mode local \
    --episode-index 0
```

## エピソードの再生

```sh
# 再生したいエピソードインデックス
export EPISODE=0

lerobot-replay \
    --robot.type=so101_follower \
    --robot.port=$FOLLOWER_PORT \
    --robot.id$FOLLOWER_ID \
    --dataset.repo_id=${HF_USER}/${REPO_NAME} \
    --dataset.episode=$EPISODE
```

## データセットの編集

[データセットツール][1]を使用して、編集できる。

エピソードの削除

```sh
# 0, 2, 5のエピソードを削除する
lerobot-edit-dataset \
    --repo_id lerobot/pusht \
    --operation.type delete_episodes \
    --operation.episode_indices "[0, 2, 5]"
```

エピソードの分割

```sh
# 80%訓練セット・20%検証セットに分割する
lerobot-edit-dataset \
    --repo_id lerobot/pusht \
    --operation.type split \
    --operation.splits '{"train": 0.8, "test": 0.2, "val": 0.2}'

# 特定のエピソードインデックスを分割する
lerobot-edit-dataset \
    --repo_id lerobot/pusht \
    --operation.type split \
    --operation.splits '{"task1": [0, 1, 2, 3], "task2": [4, 5]}'
```

データセットのマージ

```sh
# 訓練セットと検証セットを一つのデータセットに結合する
lerobot-edit-dataset \
    --repo_id lerobot/pusht_merged \
    --operation.type merge \
    --operation.repo_ids "['lerobot/pusht_train', 'lerobot/pusht_val']"
```

特徴量の削除

```sh
# カメラ特徴量を削除
lerobot-edit-dataset \
    --repo_id lerobot/pusht \
    --operation.type remove_feature \
    --operation.feature_names "['observation.images.top']"
```
