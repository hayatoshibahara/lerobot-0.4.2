# 3Dプリンターでの印刷手順

## 概要

- ロボットアーム（SO-101）の3Dプリントデータは[GitHub][1]にある
- STLファイルをダウンロードし、エンジニアカフェの3Dプリンタで印刷できる
- [オプションのアタッチメント][2]も同様に印刷が可能
    - [Overhead Camera][3]
    - [Wrist Camera][4]

[1]: https://github.com/TheRobotStudio/SO-ARM100
[2]: https://github.com/TheRobotStudio/SO-ARM100?tab=readme-ov-file#optional-hardware
[3]: https://github.com/TheRobotStudio/SO-ARM100/blob/main/Optional/Overhead_Cam_Mount_32x32_UVC_Module/README.md
[4]: https://github.com/TheRobotStudio/SO-ARM100/tree/main/Optional/SO101_Wrist_Cam_Hex-Nut_Mount_32x32_UVC_Module
