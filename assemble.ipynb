## 組み立て

公式サイトに沿って組み立てる

フォロワーアーム（白）とリーダーアーム（黒）で使用するモーターが異なる

実際に組み立てる前に、3ピンケーブルを1本取り付けておくと良い

Pro版の場合は、フォロワーアーム（白）は12V電源、リーダーアーム（黒）は5V電源を使用する

フォロワーアーム（白）で使用するモーター:

| 部位名称             |  モーター | ギア比 |
| ------------------- | :---: | :--------: |
| Base / Shoulder Pan |   1   |  1 / 345 (C001)|
| Shoulder Lift       |   2   |  1 / 345 (C001)|
| Elbow Flex          |   3   |  1 / 345 (C001)|
| Wrist Flex          |   4   |  1 / 345 (C001)|
| Wrist Roll          |   5   |  1 / 345 (C001)|
| Gripper             |   6   |  1 / 345 (C001)|

リーダーアーム（黒）で使用するモーター:

| 部位名称             |  モーター | ギア比 |
| ------------------- | :---: | :--------: |
| Base / Shoulder Pan |   1   |  1 / 191 (C044)|
| Shoulder Lift       |   2   |  1 / 345 (C001)|
| Elbow Flex          |   3   |  1 / 191 (C044)|
| Wrist Flex          |   4   |  1 / 147 (C046)|
| Wrist Roll          |   5   |  1 / 147 (C046)|
| Handle              |   6   |  1 / 147 (C046)|

### 補足：マイコンボードについて

マイコンボードは、Seed Studio XIAO用バスサーボドライバーボード

[XIAO][m-1]は、Arduinoと互換性のある小さく強力なマイコンで、ソケットに取付可能

USB（パソコンとの連携）とUART（モーターとの連携）の両方に対応している

デイジーチェーン（数珠つなぎ）で接続可能なST/SCシリーズのシリアルバスサーボ（[STS3215][m-2]）を制御できる

![](asset/motor.png)
