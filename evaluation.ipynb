# 評価

## 概要

訓練したモデルを実機で[評価][1]する。

[1]: https://huggingface.co/docs/lerobot/en/act

## 評価

初期設定後、コマンドを実行する。

```sh
# 評価結果のデータセット名（eval_から始まる必要がある）
export EVAL_DATASET_REPO_NAME="eval_grab-a-box_act_2000"

# 訓練したモデルのリポジトリ名
export MODEL_REPO_NAME="grab-a-box_act_2000"

lerobot-record \
    --robot.type=so101_follower \
    --robot.port=$FOLLOWER_PORT \
    --robot.id=$FOLLOWER_ID \
    --robot.cameras=$CAMERAS \
    --display_data=true \
    --dataset.repo_id=${HF_USER}/${EVAL_DATASET_REPO_NAME} \
    --dataset.num_episodes=$NUM_EPOSODES \
    --dataset.single_task=$PROMPT \
    --policy.path=${HF_USER}/${MODEL_REPO_NAME}
```
